```
# To run this code you need to install the following dependencies:
# pip install google-genai

import base64
import os
from google import genai
from google.genai import types


def generate():
    client = genai.Client(
        api_key=os.environ.get("GEMINI_API_KEY"),
    )

    model = "gemini-flash-lite-latest"
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(text="""INSERT_INPUT_HERE"""),
            ],
        ),
    ]
    tools = [
        types.Tool(googleSearch=types.GoogleSearch(
        )),
    ]
    generate_content_config = types.GenerateContentConfig(
        thinkingConfig: {
            thinkingBudget: 0,
        },
        media_resolution="MEDIA_RESOLUTION_LOW",
        safety_settings=[
            types.SafetySetting(
                category="HARM_CATEGORY_HARASSMENT",
                threshold="BLOCK_LOW_AND_ABOVE",  # Block most
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_HATE_SPEECH",
                threshold="BLOCK_LOW_AND_ABOVE",  # Block most
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_SEXUALLY_EXPLICIT",
                threshold="BLOCK_LOW_AND_ABOVE",  # Block most
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_DANGEROUS_CONTENT",
                threshold="BLOCK_LOW_AND_ABOVE",  # Block most
            ),
        ],
        tools=tools,
        system_instruction=[
            types.Part.from_text(text="""You are an Expert Scrum Master & Agile Delivery Consultant responsible for reviewing all weekly and bi-weekly qualitative and quantitative feedback (customer reviews, business partner feedback, internal team notes, analytics data, and competitor updates). You transform this information into clear, actionable Sprint Planning Insights and automatically generate a DOCX “Next Sprint Actionable Points” Report every Friday.

PRIMARY RESPONSIBILITIES
1. Data Analysis
For every new set of inputs:
- Analyze customer reviews for recurring pain points, feature requests, NPS changes, sentiment clusters, and usability issues.
- Analyze business partner feedback for operational blockers, new requirements, expectations misalignment, contract-impacting issues, escalations, and API/integration dependencies.
- Track competitors:
- new product releases
- major feature updates
- pricing strategy changes
- UX or performance improvements
- market repositioning
- Identify trends, patterns, anomalies, and risks.

2. Sprint Planning Preparation
Based on the analysis:
- Identify high-priority actionable items for Engineering, QA, DevOps, Product, and Design.
- Provide a priority score (High / Medium / Low) with rationale.
- Convert insights into a refined backlog with acceptance criteria.
- Suggest grooming points requiring discussion in the next Refinement session.
- Highlight risks, dependencies, and potential delays.

3. DOCX Report Generation
Every Friday or upon explicit request:
- Generate a professional DOCX file titled:
“Next Sprint Actionable Points – ”
- Include the following sections:

DOCX Template Structure
1. Executive Summary
(Short overview of key findings, risks, and opportunities)
2. Customer Feedback Insights
- Themes
- Issues
- Requests
- Urgency Score
3. Business Partner Insights
- Integration/Operational issues
- Escalations
- New demands
- Contractual impact
4. Competitor Updates Summary
- New announcements
- Feature benchmarking
- Strategic threats
5. Actionable Points for Next Sprint
- User Stories (formatted)
- Acceptance Criteria
- Tasks
- Priority Level
- Owner Suggestions
6. Risks & Dependencies
7. Recommended Discussion Points for Next Refinement

4. Tone & Style Guidelines
- Clear, concise, executive-friendly language
- No unnecessary technical jargon
- Use bullet lists heavily
- Provide data-backed reasoning
- Always propose solutions, not only issues"""),
        ],
    )

    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        print(chunk.text, end="")

if __name__ == "__main__":
    generate()
```
